# Q2

```python
df = pd.read_csv("../data/cancer patient data sets.csv")
```

```python
unique_categories = df["Level"].unique()
print("Unique categories in target variable:", unique_categories)
```

```python
print("Total Columns:", df.shape[1])
print(df.head(50))
print(df.tail(30))
df.info(memory_usage='deep')
print("Shape (rows, cols):", df.shape)
print("Number of dimensions:", df.ndim)
```

```python
num_cols = len(df.columns)
num_samples = len(df)
print("Number of columns:", num_cols)
print("Number of samples:", num_samples)
```

```python
category_counts = df["Level"].value_counts()
print(category_counts)
```

```python
sns.countplot(x="Level", data=df, palette="coolwarm")
plt.title("Distribution of Target Variable (Level)")
plt.xlabel("Level Category")
plt.ylabel("Count")
plt.show()
```

```python
subset1 = df.loc[550:900, "Gender":"Coughing of Blood"]

subset1.to_csv("subset_loc.csv", index=False)

mean_vals = subset1.mean(numeric_only=True)
median_vals = subset1.median(numeric_only=True)
mode_vals = subset1.mode(numeric_only=True).iloc[0]
std_vals = subset1.std(numeric_only=True)
min_vals = subset1.min(numeric_only=True)
max_vals = subset1.max(numeric_only=True)

print("Mean:\n", mean_vals)
print("\nMedian:\n", median_vals)
print("\nMode:\n", mode_vals)
print("\nStandard Deviation:\n", std_vals)
print("\nMinimum:\n", min_vals)
print("\nMaximum:\n", max_vals)
```

```python
subset2 = df.iloc[-20:, df.columns.get_loc("Shortness of Breath"):df.columns.get_loc("Frequent Cold")+1]

subset2.to_csv("subset_iloc.csv", index=False)

print("Mean:\n", subset2.mean(numeric_only=True))
print("Median:\n", subset2.median(numeric_only=True))
print("Mode:\n", subset2.mode(numeric_only=True).iloc[0])
print("Std Dev:\n", subset2.std(numeric_only=True))
print("Min:\n", subset2.min(numeric_only=True))
print("Max:\n", subset2.max(numeric_only=True))
```

```python
subset3 = df.loc[[110, 220, 360, 440, 656, 778, 202, 889]]

avg_alcohol = subset3["Alcohol use"].mean()
print("Average Alcohol Use for selected records:", avg_alcohol)
```

```python
print(df.head())

plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=10, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="Level", data=df, palette="viridis")
plt.title("Distribution of Target Variable (Level)")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="Level", y="Age", data=df, palette="Set2")
plt.title("Age Distribution per Level")
plt.show()
```

```python
df_encoded = df.copy()
df_encoded["Level"] = df_encoded["Level"].map({"Low": 0, "Medium": 1, "High": 2})

corr = df_encoded.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.histplot(df["Age"], kde=True, color="orange")
plt.title("Distribution of Age")

plt.subplot(1,2,2)
sns.histplot(df["Frequent Cold"], kde=True, color="green")
plt.title("Distribution of Frequent Cold")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Obesity"], color="salmon")
plt.title("Boxplot for Obesity Feature (Outlier Detection)")
plt.show()

plt.figure(figsize=(6,4))
sns.violinplot(x=df["Obesity"], color="lightgreen")
plt.title("Violin Plot for Obesity Feature")
plt.show()
```
